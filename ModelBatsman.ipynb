```python
import numpy as np # for Linear algebra
import pandas as pd # for data manipulation/CSV I/O
pd.options.mode.chained_assignment = None    # To avoid the SettingWithCopyWarning
```

```python
deliveries=pd.read_csv("dataset/deliveries.csv")
matches =pd.read_csv("dataset/matches.csv")
batsmen = pd.read_csv("dataset/batsmen.csv")
```

```python
batsmen.head()
```

```python
model_batsmen = batsmen
model_batsmen['HardHitter'] = (4 * batsmen['Fours'] + 6 * batsmen['Sixes'])/(batsmen['Balls'])
model_batsmen['Finisher'] = batsmen['NO'] / batsmen['Matches']
model_batsmen['Running'] = (batsmen['Runs'] - (4 * batsmen['Fours'] + 6 * batsmen['Sixes'])) / (batsmen['Balls'] -
                            batsmen['Fours']-batsmen['Sixes'])

model_batsmen = model_batsmen.drop(['Balls','NO','30s','50s','HS','Sixes','Fours'],axis=1)
model_batsmen.head()
```

```python
export_csv = model_batsmen.to_csv ('dataset/model_batsmen.csv', index = None, header=True) 
```
